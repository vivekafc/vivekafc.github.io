```python
#import random module

import random

feelings = ['sad', 'happy', 'angry', 'depressed']
typeof = ['annoying', 'funny', 'irritating', 'good', 'bad']
colour = ["red", "blue", "yellow", "orange", "purple", "navy", "black"]
weather = ["dull", "sunny", "rainy", "hot", "cold", "snowy"]
transport = ["bus", "train", "tram", "rickshaw", "plane", "jet"]
location = ["the mall", "the strip", "the pier", "the plaza", "the park", "the club"]
product = ["electronics", "movies", "utensils", "glassware", "records"]
purchase = ["vinyl", "phone", "laptop", "vintage copy of an old Polish film", "bong"]
food = ["creme brulee", "pizza", "pasta", "lamb bolognese", "butter chicken", "ice cream"]
cuisine = ["french", "indian", "fusion", "sweet", "british", "italian", "german"]
name = ["Earl", "Leo", "Luna", "Coco", "Eue", "Franklin"]
animal = ["turtle", "dog", "cat", "puppy", "kitten", "fish", "squirrel", "ferret"]
sleep = ["wooden mat", "mattress on the floor", "bed", "tent", "cryochamber"]

for i in range(10): 
    print("One day I woke up feeling " + random.choice(feelings) + " and I knew it was going to be a " + random.choice(typeof) + " day.")
    print("The sky was " + random.choice(colour) + " and the weather was " + random.choice(weather) +".")
    print("Outside, I took the " + random.choice(transport) + ", which took me straight to "+ random.choice(location) +".")
    print("I came across a store which sold " + random.choice(product) + " and where I managed to get the perfect " + random.choice(purchase) +".")
    print("Later, I treated myself to a "+ random.choice(food) + " at the " + random.choice(cuisine) +" restaurant.")
    print("Finally, I got back home and fed "+ random.choice(name) + " the " + random.choice(animal) + ". I curled up in my " + random.choice(sleep) + " and thought, What a Day!")
    print("\n")
```
